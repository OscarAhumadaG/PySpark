```
from pyspark.sql.functions import regexp_replace
```

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
path = '/FileStore/tables'
df = spark.read.csv(path + '/data.csv', header=True, inferSchema=True)
```

```
orders = df.select('Order ID', 'Order Date', 'Customer ID', 'Product ID', 'Quantity', 'Discount', 'Profit', 'Sales')
orders = orders.withColumn('Order ID', regexp_replace('Order ID', '-', '')) \
                .withColumn('Product ID', regexp_replace('Product ID', '-', ''))
```

```
orders.show(10)
```
